# 2. Load the Dataset

Load the dataset already we are uploaded

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load dataset

file_path = '/content/sample_data/Financial Analytics data .csv'
```

```python
df = pd.read_csv(file_path)
```

```python
# Display the first few rows of the dataset

df.head()
```

```python
df.dtypes
```

```python
df.shape
```

# 3. Data Cleaning and Preprocessing

Clean the data by handling missing values, data types, and any necessary preprocessing.

```python
# Check for missing values

df.isnull().sum()
```

```python
# Drop rows with missing values

df = df.dropna()
df
```

```python
# Convert columns to appropriate data types if necessary

df['Mar Cap - Crore'] = df['Mar Cap - Crore'].astype(float)
df['Mar Cap - Crore']
```

```python
# Convert columns to appropriate data types if necessary

df['Sales Qtr - Crore'] = df['Sales Qtr - Crore'].astype(float)
df['Sales Qtr - Crore']
```

```python
# Display cleaned data
df.info()
```

```python
# Ensure only numerical columns are used for correlation

numerical_columns = ['Mar Cap - Crore', 'Sales Qtr - Crore']
```

```python

# Summary statistics for numerical columns

df[numerical_columns].describe()
```

# 4. Exploratory Data Analysis (EDA)

Analyze the basic statistics and distribution of the data.

```python
sns.set(style = 'whitegrid')
```

```python
# Distribution of Market Cap
plt.figure(figsize=(12, 8))

sns.histplot(df['Mar Cap - Crore'], bins=50, kde=True)

plt.title('Distribution of Market Cap')
plt.xlabel('Market Cap (Crores)')
plt.ylabel('Frequency')

plt.show()
```

```python
# Distribution of Quarterly Sales
plt.figure(figsize=(10, 6))

sns.histplot(df['Sales Qtr - Crore'], bins=50, kde=True)

plt.title('Distribution of Quarterly Sales')
plt.xlabel('Quarterly Sales (Crores)')
plt.ylabel('Frequency')

plt.show()
```

# 5. Calculate Key Metrics
Identify key metrics such as average market cap and sales.

```python
# Calculate average market cap

avg_market_cap = df['Mar Cap - Crore'].mean()
print(f"Average Market Cap: {avg_market_cap:.2f} Crores")
```

```python
# Calculate and quarterly sales

avg_quarterly_sales = df['Sales Qtr - Crore'].mean()
print(f"Average Quarterly Sales: {avg_quarterly_sales:.2f} Crores")
```

# 6. Identify Relationships Between Attributes
Analyze the correlation between market cap and quarterly sales.

```python
# Select only numeric columns

numeric_df = df.select_dtypes(include=[np.number])
```

```python
# Correlation matrix

correlation_matrix = numeric_df.corr()
correlation_matrix
```

# 7. Data Visualisation

```python
# pairplot for Market Cap and Quarterly Sales
plt.figure(figsize=(12, 10))
sns.pairplot(data=df[numerical_columns])

plt.show()
```

```python

# jointplot of the correlation matrix
plt.figure(figsize=(12, 8))

sns.jointplot(data = df, x = 'Mar Cap - Crore',y = 'Sales Qtr - Crore',kind='reg',color = 'Orange')

plt.yticks(np.arange(0,110000,10000))

plt.show()
```

# 8. Conclusion and Insights

```python
print("The average market capitalization of the top 500 companies is {:.2f} crores.".format(avg_market_cap))
```

```python
print("The average quarterly sales of the top 500 companies is {:.2f} crores.".format(avg_quarterly_sales))
```

```python
print("The correlation between market cap and quarterly sales is {:.2f}.".format(correlation_matrix.loc['Mar Cap - Crore', 'Sales Qtr - Crore']))
```
